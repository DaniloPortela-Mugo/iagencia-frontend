```python
from langchain_experimental.agents.agent_toolkits import create_pandas_dataframe_agent
from langchain_openai import ChatOpenAI
import pandas as pd

df = pd.read_csv("/Users/daniloportela/Desktop/TESTES_IA/meu_app/estoque/Estoque_Veterinario.csv")
df.head(5)
```

```python
chat = ChatOpenAI(model='gpt-3.5-turbo-0125')
agent = create_pandas_dataframe_agent(
    chat,
    df,
    verbose=True,
    agent_type='tool-calling',
    allow_dangerous_code=True #Adicionando método necessário na versão 0.3 do langchain
)
```

```python
agent.invoke({'input': 'Quais produtos estão perto do vencimento e queis preciso repor no estoque?'})
```

```python
from PIL import Image
import os

# Abra a imagem base (certifique-se de que 'icone.png' existe no diretório atual)
base_image = Image.open("icone.png")

# Crie a pasta de ícones (.iconset)
iconset_folder = "MeuIcone.iconset"
os.makedirs(iconset_folder, exist_ok=True)

# Lista de tuplas com (nome do arquivo, tamanho)
icon_sizes = [
    ("icon_16x16.png", 16),
    ("icon_16x16@2x.png", 32),
    ("icon_32x32.png", 32),
    ("icon_32x32@2x.png", 64),
    ("icon_128x128.png", 128),
    ("icon_128x128@2x.png", 256),
    ("icon_256x256.png", 256),
    ("icon_256x256@2x.png", 512),
    ("icon_512x512.png", 512),
    ("icon_512x512@2x.png", 1024)
]

for filename, size in icon_sizes:
    # Redimensionar a imagem para o tamanho desejado
    resized = base_image.resize((size, size), Image.ANTIALIAS)
    # Salvar dentro da pasta .iconset
    resized.save(os.path.join(iconset_folder, filename))

print("Arquivos redimensionados gerados em:", iconset_folder)
```
